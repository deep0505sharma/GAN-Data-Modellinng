# Data Creation to csv file

```python
import regex as re
import pandas as pd
df = pd.read_json('https://raw.githubusercontent.com/selva86/datasets/master/newsgroups.json')
df2  = pd.DataFrame(columns = ['Label', 'Document'])
inde = 0
for row in df.itertuples():
    # inde+=1
    # dd=row
    data = re.sub('From.*@.*', '', row[1])
    # data = re.sub('\s+', ' ', data)
    data = re.sub("\'", "", data)
    data2 = ''
    for d in data.split('\n'):
        if 'Subject' in d:
            label = re.sub('Subject.*:\s','',d)
            label2 = re.sub(r'[^\w\s\']', ' ', label)
            continue
        else:
            data2+=d
    data3 = re.sub('[^\w\s\'.]', ' ', data2)
    # df2.append({"Label":label2,"Document":data3},ignore_index=True)
    df2.loc[len(df2.index)] =[label2,data3] #.append({'Label':label2,'Document':data3}, ignore_index = True)
    # if inde==1:
    #     break
df2.head()
    
```

```python
testdf = df2.sample(frac = 0.1)
rest_part_90 = df2.drop(testdf.index)
validdf = rest_part_90.sample(frac = 0.11)
traindf = rest_part_90.drop(validdf.index)
testdf.to_csv("test.csv")
traindf.to_csv("training.csv")
validdf.to_csv("validation.csv")
testdf.shape,traindf.shape, validdf.shape
```
